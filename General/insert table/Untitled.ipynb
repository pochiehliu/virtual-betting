```python
player = pd.read_csv('player_Oct2018.csv')
game = pd.read_csv('game_Oct2018.csv')
bet = pd.read_csv('Oct2018.csv')
```

```python
# build team first
team_list = game.away_name.unique()
team_look_up_table = {}
for idx, el in enumerate(team_list):
    print('INSERT INTO team VALUES(' + str(idx+1), ',\'' + el + '\');')
    team_look_up_table[el]=idx+1
```

```python
# build player
name_list = player.name.unique()
player_look_up_table = {}
for idx, el in enumerate(name_list):
    try: 
        first, last = el.split()
        print('INSERT INTO player VALUES(' + str(idx+1) + ',\'' + first + '\',\'' + last +'\');')
        player_look_up_table[el] = idx + 1
    except:
        print(el)
```

```python
for index, el in game.iterrows():
    print('INSERT INTO game VALUES(' + 
          el.game_id + ',' +
          el.date.replace(',','') + ',' +
          str(team_look_up_table[str(el.away_name)]) + ',' +
          str(team_look_up_table[str(el.home_name)]) + ',' +
          str(el.home_q1)+ ',' +
          str(el.home_q2)+ ',' +
          str(el.home_q3)+ ',' +
          str(el.home_q4)+ ',' +
          str(el.away_q1)+ ',' +
          str(el.away_q2)+ ',' +
          str(el.away_q3)+ ',' +
          str(el.away_q4)+ ',' +
          str(el.home_ot)+ ',' +
          str(el.away_ot) + ');'
         )
```

```python
# builg player stats
for index, el in player.iterrows():
    #print(el)
    print('INSERT INTO player_game_stats VALUES(' + 
          el.game_id + ',' +
        str(player_look_up_table[el['name']]) + ',' +
        str(team_look_up_table[el['team']]) + ',' +
        el.mp + ',' +
        str(el.fg) + ','+ # field goal made
        str(el.fga) + ','+ # field goal attempts
        str(el.tp) + ','+
        str(el.tpa) + ','+
        str(el.ft)+ ','+
        str(el.fta)+ ','+
        str(el.orb)+ ','+
        str(el.drb)+ ','+
        str(el.ast)+ ','+
        str(el.stl)+ ','+
        str(el.blk)+ ','+
        str(el.tov)+ ','+
        str(el.pf)+ ','+
        str(el.pts)+ ','+
        str(el.pm)+ ','+
        str(el.orbp)+ ','+
        str(el.drbp)+ ','+
        str(el.trbp)+ ','+
        str(el.astp)+ ','+
        str(el.stlp)+ ','+
        str(el.blkp)+ ','+
        str(el.tovp)+ ','+
        str(el.usgp)+ ','+
        str(el.ortg)+ ','+
        str(el.drtg)+ ');')
    
```
